```python
# Start

# Import necessary libraries
import cv2
from google.colab.patches import cv2_imshow

# Read the input image
image = cv2.imread('/thar.webp')

# Check if the image is successfully loaded
if image is None:
    print("Failed to load image. Please check the file path.")
else:
    # Resize the image to 300x300
    resized_img = cv2.resize(image, (500, 500))
    cv2_imshow(resized_img)
    print("Resized Image")

    # Crop a specific region (e.g., top-left 100x100 region)
    cropped_img = resized_img[0:100, 0:100]
    cv2_imshow(cropped_img)
    print("Cropped Image")

    # Convert the image to grayscale
    gray_img = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY)
    cv2_imshow(gray_img)
    print("Grayscale Image")

# End
```
